```python
import pandas as pd
import datetime
import time
```

```python
stock = 'AAPL'
period1 = int(time.mktime(datetime.datetime(2020, 12, 1, 23, 59).timetuple()))
period2 = int(time.mktime(datetime.datetime(2023, 12, 31, 23, 59).timetuple()))
interval = '1wk'
```

```python
query_string = f'https://query1.finance.yahoo.com/v7/finance/download/{stock}?period1={period1}&period2={period2}&interval={interval}&events=history&includeAdjustedClose=true'
```

```python
df = pd.read_csv(query_string)
df.to_csv()
print(df)
```
